# Further Reading

## pandas
[`pandas`](https://pandas.pydata.org/) is a Python package that is commonly used in data analysis. It has features that allow us to create and manipulate data as `DataFrame`s, then perform statistical analysis on this data. It is built on top of the `numpy` library, described below. 

## numpy
[`numpy`](https://numpy.org/doc/stable/) is a Python library that allows us to work with sets of objects more easily. It provides a multidimensional array object and various functions that perform manipulations and mathematical operations on these arrays.

## matplotlib
[`matplotlib`](https://matplotlib.org/stable/users/index) is a library used for interactive data visualization in Python. Of particular interest to us is the [`matplotlib.pyplot`](https://matplotlib.org/stable/tutorials/introductory/pyplot.html#:~:text=pyplot%20is%20a%20collection%20of,In%20matplotlib.) interface, which is used for plotting graphs.

## statsmodels
[`statsmodels`](https://www.statsmodels.org/stable/index.html) is a module used for statistical models, testing, and data analysis. In particular, we are using the [`statsmodels.formula.api`](https://www.statsmodels.org/stable/api.html) framework to create statistical models from dataframes.

## pytest
[`pytest`](https://docs.pytest.org/) is one of the most commonly used Python testing frameworks. It allows us to write functions that test whether our original code works as intended. It is particularly useful for creating small, scalable tests.

## folium
[`folium`](https://python-visualization.github.io/folium/) is a Python library used to visualize data into an interactive map.

## tweepy
[`tweepy`](https://docs.tweepy.org/en/stable/) is a Python module used to access the Twitter API.

## seaborn
[`seaborn`](https://seaborn.pydata.org/) is another Python library built on top of `matplotlib` that allows us to draw statistical graphs.

## ipywidgets
[`ipywidgets`](https://ipywidgets.readthedocs.io/en/stable/) is a library that allows us to create interactive controls for Jupyter notebooks.

## scipy
[`scipy`](https://docs.scipy.org/doc/scipy/) is a Python library built on top of `numpy` that is used for scientific computing, providing functions for optimization and statistical processing.

## geopandas
[`geopandas`](https://geopandas.org/en/stable/) is a Python module that allows us to work with geospatial data. It is built on top of `pandas` and `matplotlib`. 

## mapclassify
[`mapclassify`](https://pypi.org/project/mapclassify/) is a Python module that implements classification schemes for choropleth maps.

## xlrd
[`xlrd`](https://xlrd.readthedocs.io/en/latest/) is a library used to read data from the historical Excel `.xls` file format.

## esda
[`esda`](https://pysal.org/esda/) is a Python library used for the exploratory analysis of spatial data.

## splot
[`splot`](https://pypi.org/project/splot/) is a Python module that creates interactive visualizations for spatial data.

## libpysal
[`libpysal`](https://pysal.org/libpysal/) contains some core building blocks used in spatial analysis in python. In particular, spatial weights, input-and-output, and computational geometry.

## contextily
[`contextily`](https://contextily.readthedocs.io/en/latest/) is a Python library that allows us to retrieve tile maps from the internet. These tile maps can then be added to `matplotlib` figures.
